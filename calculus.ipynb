## 一 函数(Function)

关于函数到底是什么，我们初二就学过。所以现在我们并不会赘述它是啥。虽然高中与大学用的函数定义和初二不同，但是初中的函数定义已经够用了。

### 1.1 基本初等函数

在这一章中我们所说的所有数都是实数（除非特殊说明）

首先介绍基本初等函数。基本初等函数由以下几种函数组成：
+ 常值函数。形如 $f(x)=c$ ，其中 $c$ 是一个常数。
+ 幂函数。形如 $f(x)=x^\alpha$ ，其中， $\alpha$ 是有理数。
+ 指数函数。形如 $f(x)=a^x$ ，其中 $a$ 为实数。
+ 三角函数与反三角函数。
+ 对数函数。形如 $f(x)=\log_a(x)$ ，其中 $a$ 为实数。

常值函数的图像是一条与x轴平行的直线，实现代码如下：

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10, 0.01)
y = np.zeros_like(x) # y = 0

plt.plot(x, y)
```

幂函数由于当 $x=0$ 时 $y=0^\alpha=0$ ， $x=1$ 时 $y=1^\alpha=1$ ，因此它的图像是一条过点 $(0, 0)$ 与点 $(1, 1)$ 的直线或曲线。  
下面是函数 $y = x^2$ 的图像。

```python
%matplotlib inline

x = np.arange(-10, 10, 0.01)
y = x ** 2

plt.plot(x, y)
```

指数函数的变化非常快。它是一个恒过点 $(0, 1)$ 与点 $(1, a)$ 的曲线，其中 $a$ 是底数。下面是函数 $y=2^x$ 的图像。

```python
%matplotlib inline

x = np.arange(-3, 3, 0.01)
y = 2 ** x

plt.plot(x, y)
```

三角函数是一组（多个）函数，他们有一些特征。这里需要介绍一下。

首先我们在初三上学期学过锐角三角比。  
我们知道， $\sin 45^\circ=\frac{\sqrt2}2$ ，我们先对此进行一个改写。  
首先，将角度值化为弧度。这在微积分中至关重要（后面你会知道的）。假如 $180^\circ=\pi$ ，那么 $45^\circ=\frac\pi4$ 。因此式子变成了下面的形式：
$$\sin\frac\pi4=\frac{\sqrt2}2$$

那么，我们可以得到自变量从 $0$ 至 $\frac\pi2$ 的三角函数值。这个值在0与1之间。

现在，我们继续扩展锐角三角比。

![单位圆定义](https://img0.baidu.com/it/u=3404653646,610274295&fm=253&fmt=auto&app=138&f=PNG?w=498&h=498)

在平面直角坐标系 $xOy$ 中，以原点为圆心作一单位圆，将点 $(1,0)$ 以 $O$ 为中心逆时针旋转角度 $\theta$ 至点 $N(x, y)$ ，过点 $N$ 作 $NM$ 垂直于x轴，垂足为点 $M$ ，那么，我们设
$$\sin\theta = y,\ \cos\theta = x$$
$$\tan\theta = \frac yx,\ \cot\theta = \frac xy$$

除此之外，我们再定义两个三角函数，即正割( $\sec$ )和余割( $\csc$ )，其中：
$$\sec\theta = \frac1x$$
$$\csc\theta = \frac1y$$

需要注意的是，这些三角函数的定义方式囊括了原来我们对锐角三角比的定义。当 $0<x<\frac\pi2$ 的时候， $\triangle OMN$ 可以恰好满足我们对三角比的定义。

三角比的一些性质在三角函数中有效。

我们在锐角三角比中学过这样一些结论，它们在三角函数中依然正确：
$$\sin^2x+\cos^2x=1$$
$$\tan x\cot x=1$$
$$\sin x = \cos(\frac\pi2-x),\ \tan x = \cot(\frac\pi2-x)$$

正弦、余弦函数的图像如下图：

```python
%matplotlib inline

x = np.arange(-4, 4, 0.01)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y)
plt.plot(x, z)
```

其它三角函数由于没有正弦、余弦函数应用广泛，不予展示，可以自行搜索。

对数函数是指数函数进行逆运算。

函数 $f(x)=\log_ax$ 中，满足 $x=a^{f(x)}$ 。例如， $\log_216=4$ ， $\log_327=3$ 。  
特别地，对于两个特殊的底数，对数函数有不同表达方式。
一般，我们称 $\log_{10}x$ 为 $\lg x$，称 $\log_ex$ 为 $\ln x$ ，其中， $e\approx2.718281828$ ，以 $e$ 为底数的函数被称为自然对数，在微积分中是非常重要的存在，后面我们会介绍。  

另外注意底数为正数的对数函数，它的定义域为 $(0, +\infty)$ 。

下图是函数 $\ln x$ 的图像：

```python
%matplotlib inline

x = np.arange(0.01, 10, 0.01)
y = np.log(x)

plt.plot(x, y)
```

## 二 极限(Limit)

### 2.1 引入

我们先从一个小例子引入极限的概念。

假设你有一把测温枪，面前是滚烫的酸液，你要测量它的温度。但是很不幸，这把测温枪的说明书显示它怕酸腐蚀。那么你该如何测量？

于是你灵机一动，将测温枪对准了酸液，放在了距离它10厘米远的地方，测得温度为92摄氏度。  
你一看，效果不错，于是你继续测量，每次越来越近，下表是你的测量数据：

| 距离 | 10.0 | 9.0  | 8.0  | 7.0  | 6.0  | 5.0  | 4.0  | 3.0  | 2.0    | 1.0      |
|-----|------|------|------|------|------|------|------|------|--------|----------|
| 温度 | 92.0 | 93.9 | 94.7 | 95.6 | 96.1 | 96.3 | 96.7 | 96.9 | 96.999 | 96.99999 |

观察到，离酸液越近，你测量到的温度值越接近97摄氏度。相信你已经呼之欲出了：酸液的温度是97度。  
然而，你不能把测温枪伸进去，要是伸进去，你这把测温枪就报废了。

这种时候我们可以引入极限了。我们不妨这样说：**测温枪的读数，在距离为0处的极限为97摄氏度。**

设读数的符号为 $t$ ，距离为 $x$ ，那么我们将结论记作：
$$\lim_{x\rightarrow0}t=97℃$$

### 2.2 函数极限的计算方法

看到这个例子，你应该可以知道什么是极限，以及极限该怎么用了。
事实上，当我们遇到上面这种“自变量不在定义域上却可以在定义域无限逼近”的情况，极限确实很好用。

但是，在大多数情况下，极限的值就是函数值，可以直接计算。例如：
$$\lim_{x\rightarrow2}(x^2+2)=6$$

当然，生活中你一般遇到的极限不大可能全是直接代入就能解决的。例如下面这个极限：

$$\lim_{x\rightarrow+\infty}\frac1x$$

首先，你不可能把正无穷大代进去——正无穷是一个趋势而不是数值，但是，我们不难发现，当 $x$ 的值越来越大的时候，函数值越来越接近0。这个时候，我们就可以说：

$$\lim_{x\rightarrow+\infty}\frac1x=0$$

**一道升级版的例题** 求下式的值： $$\lim_{x\rightarrow+\infty}\frac{x+3}x$$

其实这道例题也很简单。我们知道， $\frac{x+3}x$ 可以分离出 $\frac xx$ 和 $\frac1x$ ，因此刚才的式子可以被我们变化成：
$$\frac{x+3}x=\frac3x+1$$

而我们知道， $\frac3x$ 在 $x$ 趋近于 $+\infty$ 的极限与 $\frac1x$ 的做法大同小异，也是 $0$ ，因此我们得到，上式的值为 $1+0=1$ ，因此
$$\lim_{x\rightarrow+\infty}\frac{x+3}x=1$$


**小窍门** 计算函数极限的时候，可以参考图像。

例如刚才的函数图像，你就可以明显感受到极限是 $1$ 。

```python
%matplotlib inline

x = np.arange(2, 100, 0.01)
y = (x + 3) / x

plt.plot(x, y)
```

**再来一道例题** 求 $\lim_{x\rightarrow0}\frac{\sin x}x$ 。

这道题实际上是一个结论。我们不妨先把结论写出来： $\lim_{x\rightarrow0}\frac{\sin x}x=1$ 。实际上，这道题我们可以使用几何方法证明。

外接圆半径为 $1$ 的正 $n$ 边形的面积公式为 $S=\frac n2\sin\frac{2\pi}n$ 。当 $n$ 趋近于正无穷的时候，该图形趋近于一个半径为 $1$ 的圆，面积为 $\pi$ 。

设多边形的中心角为 $x$ 。因此我们将关于 $n$ 的关系式转化为：
$$S=\frac\pi x\sin x$$

因此我们可以列出一个等式：
$$\lim_{x\rightarrow0}\frac\pi x\sin x=\pi\\\therefore\lim_{x\rightarrow0}\frac{\sin x}x=1$$

（这条结论在三角函数的微积分中应用非常广泛。）



**还有一个小窍门** 假如函数 $f(x)$ 与函数 $g(x)$ 在 $x=a$ 处的**极限存在**，那么 $\lim_{x\rightarrow a}[f(x)+g(x)]$ 等于两个极限之和。

这个窍门非常有用。对于上面的函数，当我们把 $\frac{x+3}x$ 的极限拆开的时候，本质上就用了函数 $1$ 与函数 $\frac3x$ 分别在 $x$ 趋近于 $+\infty$ 的极限之和。

好了，既然谈到了极限存在，我们就来看看，函数在某点的极限存在的条件。

### 2.3 函数极限存在的条件

我们来看下面一个例子：
$$\lim_{x\rightarrow+\infty}(x^3+4)$$

不难注意到，这个函数在x越来越大的时候，也变得越来越大，没有停下来的意思，与自变量 $x$ 一起去了正无穷。当函数在某一点趋近于无穷的时候，该函数的极限在该点不存在（D.N.E.）。

当然你可能会觉得，只有当 $x$ 在正负无穷时才会出现这种情况。其实不然，我们看看函数 $f(x)=\frac1x\ (x\in(0,+\infty))$，考虑它在0处的极限，会发现它也越来越大，趋近于正无穷。此时该函数在该点的极限也不存在。

函数在某一点存在极限的另一条件是函数在该点连续。我们以这个阶跃函数为例：
$$\begin{equation}f(x)=\left\{\begin{aligned}&y=1, x>0\\&y=0, x<0\end{aligned}\right.\end{equation}$$

如果要求 $x=0$ 处的极限，你会发现， $x=0$ 处函数并没有定义，而该点左边和右边的函数值不同。此时，我们可以记 $0$ 左端为 $0^-$ ，右端为 $0^+$ 。那么就会有：
$$\lim_{x\rightarrow0^+}=1$$
$$\lim_{x\rightarrow0^-}=0$$

出现这样的情况，即当某一点左右端求出的极限不相等时，我们称函数在该点的极限不存在。

还有一类这样的函数，我们以函数 $y=\sin\frac1x$ 为例。要求目标函数在 $x=0$ 处的极限。我们先画出函数的图像：

```python
%matplotlib inline

x = np.arange(0.001, 0.3, 0.00001)
y = np.sin(1 / x)

plt.plot(x, y)
```

可以看到，函数的图像在 $x$ 从右端越来越接近 $0$ 的时候不断加速从 $-1$ 至 $1$ 的摆动，而不是不断地趋近于一个值。在这个时候，我们也称函数在该点极限不存在。

## 三 导数(Derivative)

接下来，你将看到微积分这门学科最为精髓的部分：导数。

### 3.1 导数的定义

一个函数的导数是描述函数在某一点变化趋势的另一个函数，又名导函数值，又称微商。函数 $y=f(x)$ 的导数记作 $\frac{dy}{dx}$ 或 $\frac d{dx}f(x)$ 或 $f'(x)$。牛顿曾经称这个函数为流数——也就是函数的流动量。从定义上来说，函数在 $x=x_0$ 处的导数可以描述为，自变量 $x$ 增加无穷小量 $\Delta x$ 后，函数值的增加量 $\Delta y$ 与 $\Delta x$ 的比值。写成数学语言，就是下面这个式子：

$$\frac d{dx}f(x)=\lim_{\Delta x\rightarrow0}\frac{f(x+\Delta x)-f(x)}{\Delta x}$$

有时为了保证计算式具有对称性，还会使用这两种等价的定义方法：

$$\frac d{dx}f(x)=\lim_{\Delta x\rightarrow0}\frac{f(x+\Delta x)-f(x-\Delta x)}{2\Delta x}$$
$$\frac d{dx}f(x)=\lim_{c\rightarrow x}\frac{f(x)-f(c)}{x-c}$$

如果要通俗地理解导数，你可以认为导数就是函数某一点上取函数的**切线(tangent line)**，然后求出切线的斜率，最后将每一个点的斜率写成函数就是导数。要注意，我们只关注切线的斜率，并不关心它与坐标轴的截距是多少。如果用这种理解方式，你还能发现一个至关重要的事实（这个事实会在积分中用到）：**在定义域内一点 $x=x_0$ ，函数 $f(x)$ 与函数 $f(x)+C$ （ $C$ 为常数）的导数值相等。**

如果我们要对一个函数连续 $n$ 次计算导数，我们称结果为原函数的 $n$ 阶导（ $f^{(n)}(x)$ ）。如 $f(x)$ 的二阶导记作 $f''(x)$ 或 $\frac{d^2}{dx^2}f(x)$ 或 $f^{(2)}(x)$ ，一般来说，只要 $f$ 上面的撇写的下，我们一般都会用撇来表示，如果写不下，就用上标数字的记法。

**例题** 计算函数 $f(x)=x^2$ 的导数。

这道例题只需要用我们刚才介绍的定义式即可求解。

$$\begin{aligned}\frac d{dx}x^2&=\lim_{\Delta x\rightarrow0}\frac{(x+\Delta x)^2-x^2}{\Delta x}\\&=\lim_{\Delta x\rightarrow0}\frac{2x\Delta x+\Delta x^2}{\Delta x}\\&=\lim_{\Delta x\rightarrow0}(2x+\Delta x)\\&=2x\end{aligned}$$

看到没？就是这么简单。

当然，我们平时求导数，没有必要直接求出导数的解析式。假如我们要求出函数某点的导数，不必要求出解析式，运用计算机，我们可以轻松求出函数导数的近似值：

```python
def f(x):
    return x**2

def d(f, x0):
    dx = 1e-7
    return (f(x0 + dx) - f(x0)) / dx

print(d(f, 2))
```

我们通过函数的导数 $2x$ 可以得出，函数在该点的导数为4。此处的偏差不到 $1\times10^{-7}$ ，已经非常接近了。
利用计算机，我们还能绘制出函数导数的图像。

```python
%matplotlib inline

x = np.arange(-10, 10, 0.0001)
y = d(f, x)

plt.plot(x, y)
```

图中蓝色曲线为导数 $f'(x)=2x$ 的图像。你还可以尝试别的函数，通过这种方法得到不同函数的导数的图像。这里不再赘述。

### 3.2 导数的运算法则

#### 3.2.0 一些基本性质

之所以在3.2这一节开辟一个“第0部分”，是因为我要先介绍一些运算性质。它们对于后面的运算律应用，就像是加法之于加法运算律。

首先是我们之前所提过的，$$\frac d{dx}(f(x)+C)=\frac d{dx}f(x)$$

还有一条重要的运算性质：$$\frac d{dx}(k\cdot f(x))=k\cdot\frac d{dx}f(x)$$
理由是显然的：定义式中提取一个相同的常数 $k$ 即可。

好了，介绍完这些基本性质，我们可以正式步入本节的正题。

#### 3.2.1 积法则

积法则规定了函数之积的导数。下面是积法则的公式：设 $h(x)=f(x)g(x)$ ，那么
$$h'(x)=f'(x)g(x)+g'(x)f(x)$$

（记忆方法： $h'=f'g+fg'$ ）

那么如何证明这一法则呢？我们从导数的定义开始：

$$\begin{aligned}h'(x)&=\lim_{c\rightarrow x}\frac{f(c)g(c)-f(x)g(x)}{c-x}\\&=\lim_{c\rightarrow x}\frac{f(c)g(c)-f(x)g(x)+[f(c)g(x)-f(c)g(x)]}{c-x}\\&=\lim_{c\rightarrow x}\frac{[f(c)g(c)-f(c)g(x)]+[f(c)g(x)-f(x)g(x)]}{c-x}\\&=\lim_{c\rightarrow x}\frac{[g(c)-g(x)]f(c)+[f(c)-f(x)]g(x)}{c-x}\\&=\lim_{c\rightarrow x}[f(c)\frac{g(c)-g(x)}{c-x}+g(x)\frac{f(c)-f(x)}{c-x}]\\&=\lim_{c\rightarrow x}(f(c)g'(x)+f'(x)g(x))\end{aligned}$$

而 $\lim_{c\rightarrow x}c=x$ ，故
$$=g'(x)f(x)+f'(x)g(x)=f'(x)g(x)+g'(x)f(x)$$

#### 3.2.2 指数为整数的幂法则

看到这个名字，相信你已经知道了这个法则的使用方法。没错，这个法则规定了幂函数的导数。既然我们知道了积法则，我们不妨试一试证明指数为整数的幂法则。

为了方便你明白下面的证明过程，我们先尝试函数 $f(x)=x$ 、函数 $f(x)=x^2$ 、函数 $f(x)=x^3$ 的导数。

$$\frac d{dx}x=\lim_{\Delta x\rightarrow0}\frac{(x+\Delta x)-x}{\Delta x}=\lim_{\Delta x\rightarrow0}\frac{\Delta x}{\Delta x}=1$$
$$\frac d{dx}x^2=2x$$
$$\begin{aligned}\frac d{dx}x^3&=\lim_{\Delta x\rightarrow0}\frac{(x+\Delta x)^3-x^3}{\Delta x}\\&=\lim_{\Delta x\rightarrow0}\frac{(\Delta x)[(x+\Delta x)^2+(x+\Delta x)x+x^2]}{\Delta x}\\&=\lim_{\Delta x\rightarrow0}[(x+\Delta x)^2+(x+\Delta x)x+x^2]\\&=3x^2\end{aligned}$$

也许你眼尖的话，可能还会猜测 $f(x)=x^4$ 的导数是 $4x^3$ ，或者说，你也许猜到 $x^n$ 的导数是 $nx^{n-1}$ 。这个结论确实是对的，下面我们来证明。

假如你已经预习过了微积分，你也许知道一个运算法则叫做商法则。但是由于商法则难用（主要是难背），我们先使用另一种方法证明。

首先，我们证明 $n=0$ 的情况：

$$\frac d{dx}x^0=\lim_{\Delta x\rightarrow0}\frac{(x+\Delta x)^0-x^0}{\Delta x}=\lim_{\Delta x\rightarrow0}\frac{1-1}{\Delta x}=0=0x^{-1}\$$

然后，对于 $n=n_0\ (n_0\in\N)$ 满足条件，我们现在尝试证明 $n=n_0+1$ 的情况。

$$
\begin{aligned}
\frac d{dx}x^{n_0+1}&=\frac d{dx}(x\cdot x^{n_0})\\
&=(\frac d{dx}x)x^{n_0}+(\frac d{dx}x^{n_0})x\\
&=1\cdot x^{n_0}+nx^{n_0-1}\cdot x=(n_0+1)x^{n_0}
\end{aligned}
$$

以及 $n=-n_0$ 的情况，由于 $-n_0\leq0$ ，此时 $x\neq0$ 。

$$\begin{aligned}\because x^{-n_0}\cdot x^{n_0}&=1\\\therefore\frac d{dx}(x^{-n_0}\cdot x^{n_0})&=\frac d{dx}(1)=0\\\therefore(\frac d{dx}x^{-n_0})x^{n_0}+(\frac d{dx}x^{n_0})x^{-n_0}&=0\\\therefore(\frac d{dx}x^{-n_0})x^{n_0}+n_0x^{n_0-1}\cdot x^{-n_0}&=0\\\therefore(\frac d{dx}x^{-n_0})x^{n_0}+n_0x^{-1}&=0\\\therefore\frac d{dx}x^{-n_0}&=-n_0x^{-n_0-1}\end{aligned}$$

均符合条件。因此，对于每一个满足条件的自然数 $n_0$ ，所有大于 $n_0$ 的自然数以及其相反数均满足。而自然数 $0$ 满足条件，因此，所有整数均符合条件。

现在证明了所有整数的情况，我们来看看下面的例题：

**一道例题** 已知 $f(x)=\frac1{x^3}$ ，求 $f'(x)$ 。

这道题中函数 $\frac1{x^3}$ 可能会造成一些障碍，但是将其化为 $x^{-3}$ 就方便得多。因此答案是 $f'(x)=-3x^{-4}=-\frac3{x^4}$

**另一道例题** 已知 $f(x)=x^9$ ，求 $f^{(2022)}(x)$ 。

虽然数据 $2022$ 有一些迷惑性，但是实际上，它改成 $2021$ 或者 $100$ 对答案都没有影响。其实你可能会发现，根据幂法则的指数特点，对函数 $f(x)$ 每取一次导数，得到新函数的字母 $x$ 的指数都会减少 $1$ 。因此当我们取到 $f^{(10)}(x)$ 的时候，得到的函数是 $0$ 。要是继续取导数，只会得到 $0$ 这个答案，所以最后的答案是 $f^{(2022)}(x)=0$ 。

**一个好问题** 你证明了整数幂的幂法则的正确，那么别的呢？

利用积法则和归纳法，只能证明整数的情况。下面我们再来看另一个应用广泛的运算法则——链式法则。

#### 3.2.3 链式法则

要证明有理数幂的幂法则，我们必须借助微积分计算中一个强力的武器：**链式法则**，它规定了复合函数 $h(x)=f(g(x))$ 的导数。下面我们来看它的推导过程：

$$
\begin{aligned}
h'(x)&=\lim_{c\rightarrow x}\frac{h(x)-h(c)}{x-c}\\
&=\lim_{c\rightarrow x}\frac{f(g(x))-f(g(c))}{x-c}\\
&=\lim_{c\rightarrow x}\frac{f(g(x))-f(g(c))}{x-c}\frac{g(x)-g(c)}{g(x)-g(c)}\\
&=\lim_{c\rightarrow x}\frac{f(g(x))-f(g(c))}{g(x)-g(c)}\frac{g(x)-g(c)}{x-c}\\
&=\frac{df}{dg}\frac{dg}{dx}
\end{aligned}
$$

因此我们可以得到：

$$h'(x)=f'(g(x))g'(x)$$

#### 3.2.4 指数为有理数的幂法则

当 $n$ 为有理数时，幂法则 $\frac d{dx}x^n=nx^{n-1}$ 亦成立。

首先，我们证明 $n=\frac1q\ (q\in\Z)$ 的情况。这里，我们利用与负整数相似的证法：

$$
\begin{aligned}
\because(x^{\frac1q})^q&=x\\
\therefore\frac d{dx}(x^{\frac1q})^q&=\frac d{dx}x=1\\
\therefore{q(x^{\frac1q})^{q-1}}\cdot(\frac d{dx}x^{\frac1q})&=1\\
\therefore\frac d{dx}x^{\frac1q}&=\frac1{q(x^{\frac1q})^{q-1}}\\
\therefore\frac d{dx}x^{\frac1q}&=\frac1q(x^{-\frac1q})^{q-1}=\frac1qx^{-1+\frac1q}
\end{aligned}
$$

接着证明 $n=\frac pq\ (p,q\in\Z)$ 的情况。

$$
\begin{aligned}
\because(x^{\frac pq})&=(x^{\frac1q})^p\\
\therefore\frac d{dx}(x^{\frac pq})&=\frac d{dx}(x^{\frac1q})^p\\
\therefore\frac d{dx}(x^{\frac pq})&=p(x^{\frac1q})^{p-1}\cdot\frac1qx^{\frac1q-1}=\frac pqx^{\frac{p-1}q+\frac1q-1}\\
&=\frac pqx^{\frac pq-1}
\end{aligned}
$$

顺便说明一下，无理数的情况实际上也是成立的。由于证明原理较复杂，在这里不予叙述。

#### 3.2.5 商法则

前面其实我们提到过商法则，不过由于商法则确实难背，我们没有使用它。这里，我们对商法则进行推导：

$$
\begin{aligned}
\frac d{dx}(\frac{f(x)}{g(x)})&=\frac d{dx}(f(x)\cdot\frac1{g(x)})\\
&=\frac d{dx}(f(x)\cdot g(x)^{-1})\\
&=f'(x)\cdot g(x)^{-1}+\frac d{dx}(g(x)^{-1})\cdot f(x)\\
&=\frac{f'(x)}{g(x)}+(-g(x)^{-2})\cdot g'(x)\cdot f(x)\\
&=\frac{f'(x)\cdot g(x)-g'(x)\cdot f(x)}{g(x)^2}
\end{aligned}
$$

**温馨提示** 使用商法则的时候，要记得 $f'(x)\cdot g(x)$ 与 $g'(x)\cdot f(x)$ 的顺序。

#### 3.2.6 三角函数的导数

三角函数的导数也可以用导数的定义式推出。这里我们以函数 $\sin x$ 为例：

$$
\begin{aligned}
\frac d{dx}\sin x&=\lim_{\Delta x\rightarrow0}\frac{\sin(x+\Delta x)-\sin(x-\Delta x)}{2\Delta x}\\
&=\lim_{\Delta x\rightarrow0}\frac{(\sin x\cdot\cos\Delta x+\cos x\cdot\sin\Delta x)-(\sin x\cdot\cos\Delta x-\cos x\cdot\sin \Delta x)}{2\Delta x}\\
&=\lim_{\Delta x\rightarrow0}\frac{2\cos x\sin\Delta x}{2\Delta x}\\
&=\cos x\cdot\lim_{\Delta x\rightarrow0}\frac{\sin\Delta x}{\Delta x}\\
&=\cos x
\end{aligned}
$$

根据这一点，我们就可以得到 $\cos x$ 的导数：

$$
\begin{aligned}
\frac d{dx}\cos x&=\frac d{dx}\sin(\frac\pi2-x)\\
&=\cos(\frac\pi2-x)\cdot\frac d{dx}(\frac\pi2-x)\\
&=\sin x\cdot(-1)\\
&=-\sin x
\end{aligned}
$$

而 $\tan x=\frac{\sin x}{\cos x}$ ，因此

$$
\begin{aligned}
\frac d{dx}\tan x&=\frac d{dx}(\frac{\sin x}{\cos x})\\
&=\frac{(\frac d{dx}\sin x)\cos x-(\frac d{dx}\cos x)\sin x}{\cos^2x}\\
&=\frac{(\cos^2x+\sin^2x)}{\cos^2x}\\
&=\frac1{\cos^2x}=\sec^2x
\end{aligned}
$$

**例题** 已知 $f(x)=\sqrt{x\cos^2x+2}$ ，求 $f'(x)$ 。

这道例题要运用到的技巧涉及到前面几乎所有的运算法则。我们为了表示方便，设：

$$
\begin{aligned}
g(x)=&\sqrt x\\
h(x)=&x\cos^2x+2\\
m(x)=&x^2\\
n(x)=&\cos x\\
\end{aligned}
$$

首先，计算各函数的导数。

$$
\begin{aligned}
n'(x)=&-\sin x\\
m'(x)=&2x\\
h'(x)=&\frac d{dx}(x\cdot m(n(x))+2)\\
=&\frac d{dx}(x\cdot m(n(x)))\\
=&(\frac d{dx}x)\cdot m(n(x))+(m'(n(x))\cdot n'(x))\cdot x\\
=&1\cdot m(n(x))+x\cdot m'(n(x))\cdot n'(x)\\
=&\cos^2x+x\cdot2\cos x\cdot(-\sin x)\\
=&\cos^2x-2x\cdot\cos x\cdot\sin x\\
g'(x)=&\frac d{dx}(x^{\frac12})=\frac12x^{-\frac12}\\
\end{aligned}
$$

而 $f(x)=g(h(x))$ ，所以

$$
\begin{aligned}
f'(x)=&g'(h(x))\cdot h'(x)\\
=&\frac12(x\cos^2x+2)^{-\frac12}\cdot(\cos^2x-2x\cdot\cos x\cdot\sin x)
\end{aligned}
$$

### 3.3 导数存在的条件

上文提到过极限存在的条件。当然，导数也有存在的条件。我们不妨从定义推出导数存在的条件：

首先毋庸置疑，待求导数的这个点必须在函数的定义域内。例如 $f(x)=\sqrt x$ ，我们无法求出 $f'(-1)$ 。因为该点函数没有定义，函数的导数当然不存在。

其次，这个函数必须在该点连续。我们再次提到下面阶跃函数的例子：

$$\begin{equation}f(x)=\left\{\begin{aligned}&y=1, x\geq0\\&y=0, x<0\end{aligned}\right.\end{equation}$$

我们要求函数在 $0$ 处的导数，因此我们需要从定义出发：

$$
f'(x)=\lim_{\Delta x\rightarrow0}\frac{f(x+\Delta x)-f(x)}{\Delta x}
$$

我们先考虑 $x\rightarrow0^+$ 。此时，

$$
f'(x)=\lim_{\Delta x\rightarrow0^+}\frac{f(x+\Delta x)-f(x)}{\Delta x}=\lim_{\Delta x\rightarrow0^+}\frac{1-1}{\Delta x}=0
$$

再考虑 $x\rightarrow0^-$ 的情况：

$$
f'(x)=\lim_{\Delta x\rightarrow0^-}\frac{f(x+\Delta x)-f(x)}{\Delta x}=\lim_{\Delta x\rightarrow0^-}\frac{1-0}{\Delta x}
$$

此时极限从两个条件上不符合“存在”的定义：一方面，左右极限不相等；另一方面，右极限不存在。

因此，该函数在 $x=0$ 处的极限不存在。
故我们可以得出，函数极限存在还有一个条件，即函数在该点连续。

此处还要提到一个极为刁钻的例子，我们考虑 $f(x)=\vert x\vert$ 在 $x=0$ 处的导数。

先画出该函数的图像：

```python
%matplotlib inline

x = np.arange(-10, 10, 0.001)
y = abs(x)

plt.plot(x, y)
```

注意到，函数在 $x=0$ 处的图像就不同寻常。实际上计算过后我们会发现：函数左端导数为 $-1$ ，右端为 $1$ 。因此等同于定义式中极限不存在。此时导数不存在。

**提示** 我们可以将“函数在某点的导数不存在”的说法简写为“函数在某点不可导”。

**问题** 有没有处处不连续的函数？

答案是有的，例如狄利克雷函数 $D(x)$ 。这个函数的值的分布如下：

$$
\begin{equation}
D(x)=
\left\{
\begin{aligned}
1,\ x\in\rm{Q}\\
0,\ x\notin\rm{Q}
\end{aligned}
\right.
\end{equation}
$$

其实我们能够明显地感受到函数的不连续性。任何两个有理数之间都存在无穷多个无理数，而两个无理数之间一定能够找到有理数（这里运用了实数集的稠密性）。正是因为这一点，两个函数值为 $0$ 的点中间会有一个点函数值为 $1$ ，类似地，两个函数值为 $1$ 的点中间会有一个点函数值为 $0$ 。函数每一处都不是连续的。

**知识点** 这一类处处不连续的函数画不出图像。

**还有知识点** 由于函数处处不连续，由导数存在的条件，这个函数处处不可导。

### 4.0 指数和对数

#### 4.0.1 对数

本节将主要介绍对数的使用方法。虽然高一学对数，但是考虑到本教程需要通俗易懂，这里还是需要介绍一下对数。

对数是求幂的逆运算。例如对数的表达式 $\log_ab=m\ (a>0, b>0)$ ，一定满足 $a^m=b$ 。换句话说，底数为 $a$ 的对数就是要求出“ $b$ 是 $a$ 的几次方”。需要注意的是，对数的底必须为整数。

**例题** 求 $\log_5{125}$ 。

因为 $125=5^3$ ，所以原式 $=3$ 。

#### 4.0.2 对数的运算性质

对数有下面一系列运算性质：

**运算性质1 $\log_a(b^p)=p\log_ab$**

设 $\log_a(b^p)=m\ (m>0)$ ，因此 $a^m=b^p$ 。因为 $a>0, b>0$ ，所以 $a^{\frac mp}=b$ 。

即 $\frac mp=\log_ab$ 则 $\log_a(b^p)=p\log_ab$ 。

**运算性质2 $\log_ab+\log_ac=\log_a(bc)$**

因为 $a^{m+n}=a^ma^n$ ，所以 $m+n=\log_a(a^ma^n)$ ，设 $m=\log_ab$ ， $n=\log_ac$ 。因此 $\log_ab+\log_ac=\log_a(bc)$ 。

#### 4.0.3 特殊的对数

虽然前面已经提过一次，但是这里还是需要强调一下特殊的对数：

我们规定，底为 $10$ 的对数为 $\lg x$ ，底为 $e$ 的对数为 $\ln x$ 。

如 $\lg0.1=-1$ ， $\ln e^2=2$ 。

**例题** 求 $\lg20+\lg5$ 。

利用运算性质可以求出，原式 $=\lg(20\times5)=\lg100=2$ 。

### 4.1 常数 $e$

本教程在正式引入 $e$ 的概念之前，已经两次提到这个无理数。这一点也不奇怪，一则这个常数在微积分中的地位举足轻重，二则它太有名了。你可能很久以前就知道，有一个无理数 $e\approx2.71828183$ 。的确， $\pi$ 、 $e$ 和 $\sqrt2$ 是大众眼中最有名的三个无理数。然而其余两个有理数都在初中教材中进行了定义，我们之所以现在来讲 $e$ 的定义与性质，是因为这个数的定义本身就是一个极限。

好了，卖了这么久的关子，我还是先写出 $e$ 的定义吧：

$$e=\lim_{x\rightarrow+\infty}(1+\frac1x)^x$$

这个式子，如果当初没有因为 $e$ 而出名，也许你看到它的第一个瞬间会觉得它会往正无穷的方向跑——也就是极限不存在。但是实际上， $(1+\frac1x)^x$ 确实会不断逼近一个确切的无理数。为了帮助你理解，我把图像显示在下面：

```python
%matplotlib inline

x = np.arange(1, 50, 0.001)
y = (1 + 1/x) ** x

plt.plot(x, y)
```

可以注意到，这个函数的图像不断地向 $2.718$ 逼近。可以看到，函数的极限确实存在。

这一章主要围绕 $e$ 介绍指数、对数函数的性质。我们首先来看指数函数。

### 4.2 指数与对数函数

指数函数，作为基本初等函数之一，前面有过一段介绍。不过仅限于图像，指数函数还有很多其它的性质。

#### 4.2.1 $e^x$ 的导数

到了这里，常数 $e$ 要表现出它的第一个神奇性质了。在底数为 $e$ 的指数函数中，函数满足下面的等式：

$$\frac d{dx}(e^x)=e^x$$

你没有看错，这个函数的导数等于本身。我们来证明一下：

$$
\begin{aligned}
\frac d{dx}(e^x)=&\lim_{\Delta x\rightarrow0}\frac{e^{x+\Delta x}-e^x}{\Delta x}\\
=&e^x\cdot\lim_{\Delta x\rightarrow0}\frac{e^{\Delta x}-1}{\Delta x}\\
\end{aligned}
$$

换元，设 $t=e^{\Delta x}-1$ ，因此
$$\lim_{\Delta x\rightarrow0}t=e^0-1=0$$

又因为 $e^{\Delta x}-1=t$ ，所以
$$e^{\Delta x}=t+1\\\therefore\Delta x=\ln(t+1)$$

再换元，设 $p=\frac1t$ ，所以
$$
\begin{aligned}
\lim_{t\rightarrow0}\frac{\ln(t+1)}t
=&\lim_{p\rightarrow+\infty}\frac{\ln(\frac1p+1)}{\frac1p}\\
=&\lim_{p\rightarrow+\infty}[p\cdot{\ln(\frac1p+1)}]\\
=&\lim_{p\rightarrow+\infty}\ln(\frac1p+1)^p\\
=&\ln e=1\\
\end{aligned}
$$

对等式两边取倒数，所以
$$\lim_{t\rightarrow0}\frac t{\ln(t+1)}=1$$

再回头看原式。
$$
\begin{aligned}
\frac d{dx}(e^x)
=&e^x\cdot\lim_{\Delta x\rightarrow0}\frac{e^{\Delta x}-1}{\Delta x}\\
=&e^x\cdot\lim_{t\rightarrow0}\frac t{\ln(t+1)}\\
=&e^x\cdot1\\
=&e^x
\end{aligned}
$$

至此，我们推导完了 $f(x)=e^x$ 的导数（这也是本教程写到这里位置最长的一次推导）。下面还将介绍这个函数以及它的表亲（即一般的指数函数）的其它性质。不过在开始之前，先来看看 $e^x$ 的反函数： $\ln x$ 。

#### 4.2.2 $\ln x$ 与 $\ln\vert x\vert$

$\ln x$ 是 $e^x$ 的反函数，也许你会觉得它与 $e^x$ 同样在导数等于本身之列。然而实际上并不如此。下面，我们将推导 $\ln x$ 的导数。

$$
\begin{aligned}
\frac d{dx}(\ln x)
=&\lim_{\Delta x\rightarrow0}\frac{\ln(x+\Delta x)-\ln x}{\Delta x}\\
=&\lim_{\Delta x\rightarrow0}\frac{\ln{\frac{x+\Delta x}x}}{\Delta x}\\
=&\lim_{\Delta x\rightarrow0}\frac{\ln(1+\frac{\Delta x}x)}{\Delta x}\\
\end{aligned}
$$

换元。设 $t=\frac{\Delta x}x$ 。因此， $\Delta x=tx$ ，故有

$$
\begin{aligned}
\frac d{dx}(\ln x)
=&\lim_{t\rightarrow0}\frac{\ln(t+1)}{tx}\\
=&\frac1x\cdot\lim_{t\rightarrow0}\frac{\ln(t+1)}t\\
\end{aligned}
$$

我们证明过， $\lim_{t\rightarrow0}\frac{\ln(t+1)}t=1$ ，因此

$$
\begin{aligned}
\frac d{dx}(\ln x)
=&\frac1x\cdot1=\frac1x
\end{aligned}
$$

在这里还有一个函数需要引起我们的注意： $f(x)=\ln(-x)\ (x<0)$ 。利用链式法则，可以推出这个函数的导数为：

$$
\begin{aligned}
\frac d{dx}(\ln(-x))
=&\frac1{-x}\cdot(-1)=\frac1x
\end{aligned}
$$

因为绝对值函数 $f(x)=\vert x\vert$ 在 $x>0$ 的时候等于 $x$ ，在 $x<0$ 的时候等于 $-x$ ，因此上面两个函数可以合并为同一个函数，即：

$$f(x)=\ln\vert x\vert\ \ (x\neq0)$$

而且，由于它们的导数相同，为 $\frac1x$ ，因此，还可以得到：

$$f'(x)=\frac1x$$

**好问题** 为什么 $\ln x$ 的导数是一个幂函数？

在弄清楚这个问题之前，我们回顾一下幂法则：
$$\frac d{dx}(x^n)=nx^{n-1}$$

那么，什么样的幂函数求导能得到 $\frac1x$ 即 $x^{-1}$ ？如果单纯使用幂法则的话，得到的 $n$ 是 $0$ ，但是此时你会发现，得到的导数是 $0\cdot x^{-1}$ 。（简单理解来看，就是说系数上的 $0$ 导致了此次求导并不是 $x^{-1}$ ）

因此，我们为了看清楚为什么幂函数中间会生出这个怪胎，我们不妨对幂法则进行变形：

$$\lim_{t\rightarrow n}\frac d{dx}\frac{x^t}t=x^{n-1}$$

此处之所以使用了极限，是因为我们主要考虑的就是 $n=0$ 的情况，因此，在成功去除了系数 $0$ 的同时，还要注意待求导的函数是否有意义。因此需要在这里添加极限。那么如果 $n=0$ ，理论上（根据我们的证明），能够得到下面的等式：

$$\lim_{t\rightarrow0}\frac d{dx}\frac{x^t}t=\frac1x$$

相信你能够看得出来：假如同一定义域内两个函数的导数为同一函数，且函数有公共点，那么它们就是同一个函数——从公共点出发，沿着规定好的变化趋势走的话，必然会在同一时刻走到同一个点（就好像是两个不同的人沿着同一段人生轨迹生活一样）。我们希望看到的结论形如：

$$\lim_{t\rightarrow0}(\frac{x^t}t+C)=\ln x\ \ (x\neq0)$$

其中， $C$ 是一个待找到的常数。这里有点先猜后证的意图——我们设 $C=-\frac1t$ 。这样我们要证明的结论就是：

$$\lim_{t\rightarrow0}(\frac{x^t-1}t)=\ln x\ \ (x\neq0)$$

依然使用前面推出的结论，设 $x^t-1=p$ ，则

$$
\begin{aligned}
\lim_{p\rightarrow0}\frac p{\ln(p+1)}=&1\\
\because \ln(p+1)=\ln(x^t)=&t\ln x\\
\therefore\lim_{t\rightarrow0}\frac{x^t-1}{t\ln x}=&1\\
\therefore\lim_{t\rightarrow0}\frac{x^t-1}t=&\ln x\\
\end{aligned}
$$

这个知识点还会在积分的介绍中再次出现。
